```python
import requests
import matplotlib.pyplot as plt
```

```python
url = "http://127.0.0.1:5000/bell/15"

r = requests.get(url)
```

```python
data = r.json()
```

```python
def make_url(cnt):
    base_url = 'http://127.0.0.1:5000/bell/'
    res = base_url + str(cnt)
    return res

def get_data(cnt):
    url = make_url(cnt)
    r = requests.get(url)
    return r.json()
```

```python
cnt_list = [-5, 10, 15, 20]
for cnt in cnt_list:
    print(f'{cnt} первых чисел последовательности Белла: {get_data(cnt)}')
```

```python
y = get_data(7)
x = list(range(1, len(y)+1))
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(x, y)
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.title(f'Первые {len(y)} чисел последовательности Белаа')
plt.show()
```
